```python
child_1=0
child_0=0
Adult_1=0
Adult_0=0
Elder_1=0
Elder_0=0
count=0
for i in range(df1.shape[0]):
    if df1['Survived'][i] == 1 & df1['Age'][i]==0:
        child_1+=1
    elif df1['Survived'][i] == 0 & df1['Age'][i]==0:
        child_0+=1
    elif df1['Survived'][i] == 1 & df1['Age'][i]==1:
        Adult_1+=1
    elif df1['Survived'][i] == 0 & df1['Age'][i]==1:
        Adult_0+=1
    elif df1['Survived'][i] == 1 & df1['Age'][i]==2:
        Elder_1+=1
    elif df1['Survived'][i] == 0 & df1['Age'][i]==2:
        Elder_0+=1
    else:  
        print(df1['Age'][i],df1['Survived'][i])
print("Number of Children Survived",child_1)
print("Number of Children Not Survived",child_0)
print("Number of Adults Survived",Adult_1)
print("Number of Adults Not Survived",Adult_0)
print("Number of Elderly Survived",Elder_1)
print("Number of Elderly Not Survived",Elder_0)
print("Number of rows where age is not 0,1 or 2",count)
```
